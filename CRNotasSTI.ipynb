Solução Pandas

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/notas.csv")
```

```python
medias_por_aluno = df.groupby('MATRICULA')['NOTA'].agg(['mean'])
medias_por_aluno = medias_por_aluno.round(2)


print("Estatísticas por aluno:")
print("-" * 50)
print(medias_por_aluno)
```

```python
medias_por_curso = df.groupby('COD_CURSO')['NOTA'].agg(['mean'])
medias_por_curso = medias_por_curso.round(2)

print("Estatísticas por curso:")
print("-" * 50)
print(medias_por_curso)
```

Solução orientada a objetos

Classe aluno

```python
class aluno:
  #necessário declarar notas como lista no init para facilitar o uso da
  #leitura do pandas
  def __init__(self, matricula,notas):
    self.matricula = matricula
    self.notas = list(notas)

  def add_nota(self, nota):
    self.notas.append(nota)

  def get_matricula(self):
    return self.matricula

  def get_media(self):
    return sum(self.notas)/len(self.notas)
```

Classe Curso

```python
class curso:
  def __init__(self,cod_curso,notas):
    self.cod_curso = cod_curso
    self.notas = list(notas)

  def add_nota(self, nota):
    self.notas.append(nota)

  def get_cod_curso(self):
    return self.cod_curso

  def get_media(self):
    return sum(self.notas)/len(self.notas)
```

```python
teste_al = aluno(50,[])

teste_al.add_nota(80)
teste_al.add_nota(64)
teste_al.add_nota(77)

teste_cur = curso(200,[])

teste_cur.add_nota(70)
teste_cur.add_nota(80)
teste_cur.add_nota(90)

print(teste_al.get_matricula())
print(teste_al.get_media())
print("\n-----\n")
print(teste_cur.get_cod_curso())
print(teste_cur.get_media())
```

Gerenciador de notas

```python
class gerenciador:
  def __init__(self):
    self.alunos = {}
    self.cursos = {}

  def ler_arquivo_alunos(self,arquivo):
    df = pd.read_csv(arquivo)
    for mat, grupo in df.groupby('MATRICULA')['NOTA']:
      self.alunos[mat] = aluno(mat, grupo)

  def mostrar_media_alunos(self):
    print("Média dos alunos:\n")
    for aluno in self.alunos.values():
      print(aluno.get_matricula(),": ", aluno.get_media())

  def ler_arquivo_cod_curso(self,arquivo):
    df = pd.read_csv(arquivo)
    for cod, grupo in df.groupby('COD_CURSO')['NOTA']:
      self.cursos[cod] = curso(cod, grupo)

  def mostrar_media_cursos(self):
    print("Média dos cursos:\n")
    for curso in self.cursos.values():
      print(curso.get_cod_curso(),":", curso.get_media())
```

```python
teste_arq = gerenciador()
teste_arq.ler_arquivo_alunos("/content/drive/MyDrive/notas.csv")
teste_arq.mostrar_media_alunos()
print("\n-----\n")
teste_arq.ler_arquivo_cod_curso("/content/drive/MyDrive/notas.csv")
teste_arq.mostrar_media_cursos()

```
